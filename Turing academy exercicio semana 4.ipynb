```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('Pokemon.csv')
df.drop_duplicates()
cor_1 = 'lightgray'
cor_2 = 'gold'
labels = ['Comum','Lendário']
plt.figure(figsize = (10,6))
sns.countplot(x='Generation', hue='Legendary', data=df, palette = [cor_1,cor_2])
plt.legend(labels = labels)
plt.title ('Distribuição de Pokémons Lendários e comuns por geração')
plt.xlabel('Geração')
plt.ylabel('Quantia de Pokémons')
plt.show()
```

```python
quantia = df.groupby(['Generation'])['Generation'].count()
quantia_acumulada = []
soma = 0
for numero in quantia:
  soma = soma + numero
  quantia_acumulada = quantia_acumulada + [soma]
quantia_acumulada
```

```python
geracao_quantia = np.arange(1,7)
sns.lineplot(x = geracao_quantia, y = quantia_acumulada, color = 'red')
plt.xlabel('geração')
plt.ylabel('Quantia total de Pokémons')
plt.title('Evolução da quantia total de Pokémons ao decorrer das gerações')

```

```python
plt.figure(figsize = (15,6))
cores = ['darkgreen', 'red', 'blue', 'lightgreen','gray','purple','lightyellow','brown','lightpink','darkred','purple','brown','darkgray','lightblue','darkblue','black','lightgray','gold']
sns.countplot( x =  'Type 1', hue = 'Type 1', data = df, palette = cores)
plt.xlabel('tipo principal')
plt.ylabel('')
```
